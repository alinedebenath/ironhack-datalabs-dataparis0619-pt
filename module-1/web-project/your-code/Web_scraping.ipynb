```python
import requests
from bs4 import BeautifulSoup
import re
# Scrapin the wiki page of Python (mythology)
url = 'https://en.wikipedia.org/wiki/Python_(mythology)'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
data = soup.find_all('a',{'class': None})
gods = [d.get('title') for d in data]
gods
```

```python
# Searching the Greek Gods, so the one word items..
greeks = []
for g in gods:
    if g:
        z = re.findall('^[a-zA-Z0-9]*$', g)
        greeks.append(z)
greek_gods = sum(greeks, [])
# Cleaning what remains
greek_gods.remove('Wikisource')
greek_gods.remove('Sacrilege')
print(greek_gods)
```

```python
with open("Web_scraping_output.txt", "w") as output:
    output.write(str(greek_gods))
```
