```python
import pandas as pd
import plotly as py
import cufflinks as cf
from ipywidgets import interact

cf.go_offline()
```

```python
data = pd.read_excel('../data/Online Retail.xlsx')
data.head()
```

## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
import datetime
a = data.loc[(data['Country'] !='United Kingdom')&(data['InvoiceDate'].dt.year == 2011)&(data['InvoiceDate'].dt.month == 4), ['Quantity', 'Country','Revenue']]
a.iplot(kind='bar', x='Country')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
b = data.loc[(data['Country'] =='France')&(data['InvoiceDate'].dt.year == 2011)&(data['InvoiceDate'].dt.month >= 4)&(data['InvoiceDate'].dt.month <= 5), ['Quantity', 'Country','Revenue']]
b.iplot(kind='bar', x='Country')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python

c = data.loc[(data['Description'].str.contains('PARTY BUNTING'))].groupby('Country', as_index=False).mean()

c.iplot(x='Quantity', y='UnitPrice', categories='Country',
        xTitle ='Average Quantity',yTitle='Average Unit Price',
          title='Average Quantity vs. Average Unit Price by Country')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
d = data.loc[(data['Country'] =='EIRE')|(data['Country']=='Germany')|(data['Country'] =='France')|(data['Country'] =='Netherlands')]
d = d.pivot_table(values='Quantity', columns='Country', index='InvoiceNo', aggfunc='sum')
d.iplot(kind='hist', xTitle='InvoiceNo', subplots=True, yTitle='Quantity', title='Distributions of quantity per invoice')
```
