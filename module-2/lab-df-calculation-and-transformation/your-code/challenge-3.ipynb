## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
amounts = orders.groupby('CustomerID').sum()
amounts.head()
```

```python
quantile95 = amounts['amount_spent'].quantile(0.95)
quantile75 = amounts['amount_spent'].quantile(0.75)
```

```python
amounts_95 = amounts[amounts['amount_spent']>quantile95]
amounts_75 = amounts[(amounts['amount_spent']>quantile75) & (amounts['amount_spent']<quantile95)]  
amounts_75
```

```python
#### your code here
#amounts.insert(9,'Status','')
amounts.loc[amounts_95.index, 'Status'] = 'VIP'
amounts.loc[amounts_75.index,'Status'] = 'Prefered Client'
amounts.head()
```

# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
# your code here
orders_new = orders.merge(amounts['Status'], left_on = 'CustomerID', right_index = True)
orders_new['Country'].mode()
```

```python
VIP_country = orders_new[orders_new['Status'] == 'VIP'].groupby('Country').count()
PC_country = orders_new[(orders_new['Status'] == 'Prefered Client')|(orders_new['Status'] == 'VIP')].groupby('Country').count()
print('VIP =', VIP_country.index.max())
print('Prefered Clients & VIP =', PC_country.index.max())
```
